# Taking a Look at Source Code

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link = 'https://packtpub.com'
```

```python
def show_html(link):
    soup = bs(requests.get(link).text, 'html.parser')
    for js in soup('script'):
        js.decompose()
    return soup.prettify()
```

```python
print(show_html(link))
```
