# Extracting Links

```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
link ='https://packtpub.com'
```

```python
soup = bs(requests.get(link).text, 'html.parser')
```

```python
links = soup.find_all('a')
```

```python
type(links)
```

```python
len(list(links))
```

```python
for link in links:
    print(link.text)
```
